```python
# ================================================
#   Google Colab：從 Google Drive 讀取檔案並批量重新命名
# ================================================

from google.colab import drive
import os
import shutil
new_names = {}
# 步驟 1：掛載 Google Drive
print("正在掛載 Google Drive...")
drive.mount('/content/drive')

# 步驟 2：指定你的 Google Drive 資料夾路徑
# 範例：'/content/drive/MyDrive/要改名的檔案'
input_path = input("請輸入 Google Drive 要改名的檔案 (/content/drive/MyDrive/要改名的檔案）： ").strip()
folder_path = "/content/drive/MyDrive/" + input_path

if not os.path.isdir(folder_path):
    print(f"路徑不存在：{folder_path}")
    print("請檢查路徑是否正確/先在 GoogleDrive 上傳檔")
else:
    # 取得資料夾內所有檔案（不包含子資料夾）
    all_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    if not all_files:
        print("這個資料夾沒有檔案。")
    else:
        print(f"\n資料夾內找到 {len(all_files)} 個檔案：")
        for f in all_files:
          print(f)

        # 讓使用者選擇要改名的檔案（全部或部分）
        print(f"\n先將全部file改至file_1...file{len(all_files)}")
        for i, old in enumerate(all_files, 1):
          base, ext = os.path.splitext(old)
          new = f"file{i:03d}{ext}"
          new_names[old] = new
        renamed_count = 0
        for old, new in new_names.items():
            if old != new:
                old_path = os.path.join(folder_path, old)
                new_path = os.path.join(folder_path, new)
                try:
                    os.rename(old_path, new_path)
                    renamed_count += 1
                    print(f"成功：{old} → {new}")
                except Exception as e:
                    print(f"失敗：{old} → {new} ({e})")
        go_or_not = "yes"
        while go_or_not == "yes":
          target_start_number = input("你想要改名file的 starting number:(eg:1)")
          target_end_number = input("你想要改名file的 ending number:(eg:9)")
          mj_type = input("你想改邊種麻雀：(筒子：tung/萬子：mang/索子：suo/番子：fan)").strip()
          mj_number = input("你想改第幾隻(eg:1,2,3...)").strip()
          selected_files = []
          name_lsit =[]   #for further training model to read the files
          for i in range(int(target_start_number), int(target_end_number )+ 1):
            folder_name = f"file{i:03d}.jpg"
            selected_files.append(folder_name)
          for i, old in enumerate(selected_files, 1):
                      base, ext = os.path.splitext(old)
                      new = f"{mj_number}{mj_type}{i:03d}{ext}"
                      new_names[old] = new
          renamed_count = 0
          for old, new in new_names.items():
              if old != new:
                  old_path = os.path.join(folder_path, old)
                  new_path = os.path.join(folder_path, new)
                  try:
                      os.rename(old_path, new_path)
                      renamed_count += 1
                      print(f"成功：{old} → {new}")
                  except Exception as e:
                      print(f"失敗：{old} → {new} ({e})")
          go_or_not = input ("繼續嗎(yes/no)")
mj_name_lis = new_names.keys()


```
