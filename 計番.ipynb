```python
# from itertools import permutations

hand = [
    "1tung", "2tung", "nam",
    "1tung", "1tung", "nam",
    "3tung", "3tung", "nam",
    "dong", "dong", "dong",
    "3tung", "3tung"]

hand_copy = hand.copy()  # 整一個copy for 扣左啲計左番既牌
round = input("請輸入現在是什麼圈(東：dong/南：nam/西：sike/北：bulk)")
position = input("請輸入食糊的玩家是在咩位置(東：dong/南：nam/西：sike/北：bulk)")
how_to_eat = input("請輸入:怎樣食(打出：1 / 自摸：2)")

def count_hand(hand_copy):
    count_tungji = {f"{i}tung": 0 for i in range(1, 10)}
    count_mangji = {f"{i}mang": 0 for i in range(1, 10)}
    count_suoji = {f"{i}suo": 0 for i in range(1, 10)}
    count_fanji = {"dong": 0,"nam":0,"sike":0,"buck":0,"chung":0,"fat":0,"white_board":0}


    for tile in hand_copy:  # tile --> 每張手牌
        if tile in count_tungji:  # 檢查目前這張牌（tile）是否出現在 count 這個字典的 key 裡
          count_tungji[tile] += 1  # 如果 tile 是筒子牌就把對應的計數 加 1

        if tile in count_mangji:
          count_mangji[tile] += 1

        if tile in count_suoji:
          count_suoji[tile] +=1

        if  tile in count_fanji:
          count_fanji[tile] +=1


    return count_tungji,count_mangji,count_suoji,count_fanji
def find_type(count_tungji,count_mangji,count_suoji,count_fanji):
  hv_tungji_or_not = False #牌種數類
  hv_mangji_or_not = False
  hv_suoji_or_not = False
  hv_fanji_or_not = False
  for value in count_tungji.values():
      if value != 0:
        hv_tungji_or_not = True
  for value in count_mangji.values():
      if value != 0:
        hv_mangji_or_not = True
  for value in count_suoji.values():
      if value != 0:
        hv_suoji_or_not = True
  for value in count_fanji.values():
        if value != 0:
          hv_fanji_or_not = True
  return hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not





def show_hand(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji):
    # Print result
    print("手牌：")
    for num in range(1, 10):  # num:1-9
        tungji = f"{num}tung"  # 1tung,2tung...
        mangji = f"{num}mang"  # 1mang,2mang...
        suoji = f"{num}suo"    # 1suo,2suo...
        print(f"{tungji}: {count_tungji[tungji]} 隻")  # count[tile_name] -->數量
        print(f"{mangji}: {count_mangji[mangji]} 隻")  # count[tile_name] -->數量
        print(f"{suoji}: {count_suoji[suoji]} 隻")  # count[tile_name] -->數量
    fanji_lis = ["dong","nam","sike","buck","chung","fat","white_board"]
    for fanji in fanji_lis:
      print(f"{fanji}: {count_fanji[fanji]} 隻")


# # #Check眼
def detect_eyes(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji):
    eyes = 0
    eyes_lis = []
    for num in range(1, 10):
        tungji = f"{num}tung"
        mangji = f"{num}mang"  # 1mang,2mang...
        suoji = f"{num}suo"
        if count_tungji[tungji] >= 2:
            eyes += 1
            eyes_lis.append(tungji)
        if count_mangji[mangji] >=2:   #use if instead of elif coz it can be fuilfill both condition at the same times
          eyes += 1
          eyes_lis.append(mangji)
        if count_suoji[suoji] >=2:
          eyes += 1
          eyes_lis.append(suoji)
    fanji_lis = ["dong","nam","sike","buck","chung","fat","white_board"]
    for fanji in fanji_lis:
      if count_fanji[fanji] >= 2:
            eyes += 1
            eyes_lis.append(fanji)


    print(f"{eyes_lis} 有兩隻或以上--＞可以做眼")
    return eyes_lis, eyes


def del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji):
    print(f"用{eyes_real}做眼")
    hand_calculate = hand_copy.copy()
    for i in range(2):
        hand_calculate.remove(eyes_real)
        if eyes_real in count_tungji:
          count_tungji[eyes_real] -= 1
        elif eyes_real in count_mangji:
          count_mangji[eyes_real] -= 1
        elif eyes_real in count_suoji:
          count_suoji[eyes_real] -= 1
        elif eyes_real in count_fanji:
          count_fanji[eyes_real] -= 1
    return hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji


# # Check aaa 3隻一樣 , m -->數量
def detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji):
    m = 0
    print("碰:")
    for num in range(1, 10):
        tungji = f"{num}tung"
        mangji = f"{num}mang"  # 1mang,2mang...
        suoji = f"{num}suo"
        if count_tungji[tungji] >= 3:
            print(f"{tungji}有3張")
            for i in range(3):
                hand_calculate.remove(tungji)
                count_tungji[tungji] -= 1
            m = m + 1
        if count_mangji[mangji] >=3:   #use if instead of elif coz it can be fuilfill both condition at the same times
          print(f"{mangji}有3張")
          for i in range(3):
            hand_calculate.remove(mangji)
            count_mangji[mangji] -= 1
          m = m + 1
        if count_suoji[suoji] >=3:
          print(f"{suoji}有3張")
          for i in range(3):
            hand_calculate.remove(suoji)
            count_suoji[suoji] -= 1
          m = m + 1
    fanji_lis = ["dong","nam","sike","buck","chung","fat","white_board"]
    for fanji in fanji_lis:
      if count_fanji[fanji] >=3:
        print(f"{fanji}有3張")
        for i in range(3):
              hand_calculate.remove(fanji)
              count_fanji[fanji] -= 1
        m = m + 1
    print("淨低既牌：", hand_calculate)
    return m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji


# # Check bbbb 4隻一樣(槓) , p -->數量
def detect_BBBB(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji):
    p = 0
    print("槓:")
    for num in range(1, 10):
        tungji = f"{num}tung"
        mangji = f"{num}mang"  # 1mang,2mang...
        suoji = f"{num}suo"
        if count_tungji[tungji] == 4:
            print(f"{tungji}有4張")
            for i in range(4):
                hand_calculate.remove(tungji)
                count_tungji[tungji] -= 1
            p =+ 1
        if count_mangji[mangji] ==4 :   #use if instead of elif coz it can be fuilfill both condition at the same times
          print(f"{mangji}有4張")
          for i in range(4):
            hand_calculate.remove(mangji)
            count_mangji[mangji] -= 1
          p += 1
        if count_suoji[suoji] == 4:
          print(f"{suoji}有4張")
          for i in range(4):
            hand_calculate.remove(suoji)
            count_suoji[suoji] -= 1
          p += 1
    fanji_lis = ["dong","nam","sike","buck","chung","fat","white_board"]
    for fanji in fanji_lis:
      if count_fanji[fanji] == 4:
        print(f"{fanji}有4張")
        for i in range(4):
              hand_calculate.remove(fanji)
              count_fanji[fanji] -= 1
        p += 1
    print("淨低既牌：", hand_calculate)
    return p, count_tungji, hand_calculate,count_mangji,count_suoji,count_fanji


# Check abc 3隻順住 , n -->數量
def detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji):
    how_many = 0
    n = 0
    print("3隻順住")
    for num in range(1, 8):
        tile_name = f"{num}tung"
        next_one_tile_name = f"{num+1}tung"
        next_two_tile_name = f"{num+2}tung"
        how_many = min(
            count_tungji[tile_name],
            count_tungji[next_one_tile_name],
            count_tungji[next_two_tile_name])
        if how_many > 0:
            for i in range(how_many):
                hand_calculate.remove(tile_name)
                hand_calculate.remove(next_one_tile_name)
                hand_calculate.remove(next_two_tile_name)
                count_tungji[tile_name] -= 1
                count_tungji[next_one_tile_name] -= 1
                count_tungji[next_two_tile_name] -= 1
                n += 1
            print(f"順子：{tile_name} + {next_one_tile_name} + {next_two_tile_name} ,數量：{how_many}")
    for num in range(1, 8):
        tile_name = f"{num}mang"
        next_one_tile_name = f"{num+1}mang"
        next_two_tile_name = f"{num+2}mang"
        how_many = min(
            count_mangji[tile_name],
            count_mangji[next_one_tile_name],
            count_mangji[next_two_tile_name])
        if how_many > 0:
            for i in range(how_many):
                hand_calculate.remove(tile_name)
                hand_calculate.remove(next_one_tile_name)
                hand_calculate.remove(next_two_tile_name)
                count_mangji[tile_name] -= 1
                count_mangji[next_one_tile_name] -= 1
                count_mangji[next_two_tile_name] -= 1
                n += 1
            print(f"順子：{tile_name} + {next_one_tile_name} + {next_two_tile_name} ,數量：{how_many}")

    for num in range(1, 8):
        tile_name = f"{num}suo"
        next_one_tile_name = f"{num+1}suo"
        next_two_tile_name = f"{num+2}suo"
        how_many = min(
            count_suoji[tile_name],
            count_suoji[next_one_tile_name],
            count_suoji[next_two_tile_name])
        if how_many > 0:
            for i in range(how_many):
                hand_calculate.remove(tile_name)
                hand_calculate.remove(next_one_tile_name)
                hand_calculate.remove(next_two_tile_name)
                count_suoji[tile_name] -= 1
                count_suoji[next_one_tile_name] -= 1
                count_suoji[next_two_tile_name] -= 1
                n += 1
            print(f"順子：{tile_name} + {next_one_tile_name} + {next_two_tile_name} ,數量：{how_many}")

    print("淨低既牌：", hand_calculate)
    return n, count_tungji, hand_calculate,count_mangji,count_suoji

def check_eat(n, m, p, eyes, hand_calculate,):
    print(f"M:{m} N:{n} P:{p}")
    eat_or_not = 0
    if (n + m + p) == 4 and eyes > 0 and not hand_calculate:
        print("這是一個糊")
        eat_or_not += 1
    else:
        print("這不是一個糊")
    return eat_or_not,n, m, p

def first_AAA_next_ABC(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji):
    hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
    m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
    p = 0
    eat_or_not,n, m, p = check_eat(n, m, p, eyes, hand_calculate)
    print("###########################")
    return eat_or_not,n, m, p

def first_ABC_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji):
    hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
    n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
    m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    p = 0
    eat_or_not,n, m, p = check_eat(n, m, p, eyes, hand_calculate)
    print("###########################")
    return eat_or_not,n, m, p

def firstBBBB_next_ABC_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji):  # 槓上碰(GSP)
    hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
    p, count_tungji, hand_calculate,count_mangji,count_suoji,count_fanji = detect_BBBB(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
    m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    eat_or_not,n, m, p = check_eat(n, m, p, eyes, hand_calculate)
    print("###########################")
    return eat_or_not,n, m, p

def firstBBBB_next_AAA_next_ABC(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji):  # 槓碰上(GPS)
    hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
    p, count_tungji, hand_calculate,count_mangji,count_suoji,count_fanji = detect_BBBB(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
    eat_or_not,n, m, p = check_eat(n, m, p, eyes, hand_calculate)
    print("###########################")
    return eat_or_not,n, m, p

def firstABC_next_BBBB_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji):  # 上槓碰(SGP)
    hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
    n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
    p, count_tungji, hand_calculate,count_mangji,count_suoji,count_fanji = detect_BBBB(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
    eat_or_not,n, m, p = check_eat(n, m, p, eyes, hand_calculate)
    print("###########################")
    return eat_or_not,n, m, p
def calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat):
  found = 0
  found_name_lis = []
  if n == 4 and m == 0 and p == 0 :
    print("平糊")
    found += 1
  elif  n == 0 and (m+p) == 4 :
    print("對對糊")
    found += 3
  elif n != 0 and (m != 0 or p != 0 ):
    print("雞糊")
  count_type = [hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ].count(True)
  if count_type == 1:
    if hv_fanji_or_not == True:
      print("混一色")
      found += 3
    else:
      print("清一色")
      found += 7
  count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #計番子既番
  chung_fat_ba_lis = ["chung","fat","white_board"]
  dnsb_lis = ["dong","nam","sike","buck"]
  for fanji in chung_fat_ba_lis:
    if count_fanji[fanji] >=3:
      found +=1
  for fanji in dnsb_lis:
    if count_fanji[fanji] >=3:
      if position == fanji and round == fanji:
        found +=2   #eg. 雙番東
        print(f"雙番{fanji}")
      elif position == fanji:
        found +=1
        print(f"{fanji}有番")
  if  how_to_eat == 2:
    print("自摸")
    found +=1





  print(f"{found}番")

  return found









found_lis = []
count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy)
hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not = find_type(count_tungji,count_mangji,count_suoji,count_fanji)
show_hand(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
print("#####################################################################")
eyes_lis, eyes = detect_eyes(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
print("#####################################################################")

for eyes_real in eyes_lis:
    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count  # need use hand.copy instead of hand because it may change the value of hand
    eat_or_not,n, m, p = first_AAA_next_ABC(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)  # 碰上 (PS)
    if eat_or_not == 1:
        print(f"食糊：用{eyes_real}做眼--＞碰--＞上 ")
        found_lis .append (calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat))

    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
    eat_or_not,n, m, p = first_ABC_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)  # 上碰 (SP)
    if eat_or_not == 1:
        print(f"食糊：用{eyes_real}做眼--＞上--＞碰 ")
        found_lis .append (calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat))

    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
    eat_or_not,n, m, p = firstBBBB_next_ABC_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)  # 槓上碰(GSP)
    if eat_or_not == 1:
        print(f"食糊：用{eyes_real}做眼--＞槓-->上-->碰")
        found_lis .append (calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat))

    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
    eat_or_not,n, m, p = firstBBBB_next_AAA_next_ABC(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)  # 槓碰上(GPS)
    if eat_or_not == 1:
        print("fuck",count_fanji)
        print(f"食糊：用{eyes_real}做眼--＞槓-->碰-->上 ")
        found_lis .append (calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat))

    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
    eat_or_not,n, m, p = firstABC_next_BBBB_next_AAA(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)  # 上槓碰(SGP)
    if eat_or_not == 1:
        print(f"食糊：用{eyes_real}做眼--＞上--＞槓-->碰 ")
        found_lis .append (calculate_found(n, m, p,hv_tungji_or_not ,hv_mangji_or_not ,hv_suoji_or_not ,hv_fanji_or_not,hand_copy,round,position,how_to_eat))
    count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
print(f"{max(found_lis)}番")
# count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy)
# show_hand(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
# print("#####################################################################")
# eyes_lis, eyes = detect_eyes(hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
# print("#####################################################################")
# for eyes_real in eyes_lis:
  #  hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji = del_eyes(eyes_real, hand_copy, count_tungji,count_mangji,count_suoji,count_fanji)
  #  m,  hand_calculate,count_tungji,count_mangji,count_suoji,count_fanji = detect_AAA(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
  #  p, count_tungji, hand_calculate,count_mangji,count_suoji,count_fanji = detect_BBBB(hand_calculate, count_tungji,count_mangji,count_suoji,count_fanji)
  #  n, count_tungji, hand_calculate,count_mangji,count_suoji = detect_ABC(hand_calculate, count_tungji,count_mangji,count_suoji)
  #  eat_or_not = check_eat(n, m, p, eyes, hand_calculate)
  # first_AAA_next_ABC(count_tungji, eyes_real,hand_copy,count_mangji,count_suoji,count_fanji)
  # count_tungji,count_mangji,count_suoji,count_fanji = count_hand(hand_copy) #reset the count
```
