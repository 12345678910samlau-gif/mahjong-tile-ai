```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,2,figsize =(10,8)) #build a subplot
url = "https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv" #insert the csv file
data = pd.read_csv(url)
x = data['YearsExperience']
y = data['Salary']
b = 0
w = 10
#plot個 with b= 0 & W = 10 graph
ax[0].scatter(x,y,marker='x',color='red')
ax[0].set_xlabel ( 'Years Experience')
ax[0].set_ylabel ( 'Salary')
y_pred = w*x+b
ax[0].plot(x,y_pred)
ax[0].set_title('Year Experience X Salary')
#Plot 個 成本函數 Cost Function (B = 0, W = -100 TO 100 )
cost2 = []
for w in range (-100,101):
 cost = (y - (w*x + b))**2
 cost = cost.sum()/len (x)
 cost2.append(cost)
w = np.arange(-100,101)
ax[1].plot(w,cost2)
ax[1].set_title('Cost Function (B = 0, W = -100 TO 100')
ax[1].set_xlabel ( 'W')
ax[1].set_ylabel ( 'cost')



```
